```python
import pandas as pd
import spacy
nlp = spacy.load('pl_core_news_lg')
```

```python
from spacy.lang.pl.stop_words import STOP_WORDS
```

```python
df = pd.read_csv('/content/drive/MyDrive/dane7.csv')
df
```

```python
#6. Usuń wszystkie słowa oznaczone jako stopwords i przekaż oczyszczony tekst do nowej kolumny o nazwie text_without_stopwords
def remove_stopwords_spacy(text):
  if pd.isna(text):
    return ''
  doc = nlp(text)
  filtrowane = [token.text for token in doc if not token.is_stop]
  return ' '.join(filtrowane)
df['text_without_stopwords'] = df['text'].apply(remove_stopwords_spacy)
```

```python
#Część 2
#1. Wypisz top5 tweetów z największa liczbą polubień.
top5= df.nlargest(5, 'favorite_count')[['text', 'favorite_count']]
top5
```

```python
#2. Wypisz top5 tweetów z największa liczbą retweetów
top5_retweet= df.nlargest(5, 'retweet_count')[['text', 'retweet_count']]
top5_retweet
```

```python
#3. Wyświetl tylko te tweety, które nie są uznane jako “wrażliwe” (kolumna possibly_sensitive).
nwrazliwe_tweety= df[df['possibly_sensitive'] == False]
nwrazliwe_tweety
```

```python
#4. Wyświetl tweety użytkownika, który założył konto najwcześniej (spośród wszystkich użytkowników w datasecie) (kolumna user_created_at).
wczesny_user = df.loc[df['user_created_at'].idxmin()]
wczesny_user_tweety = df[df['user_id'] == wczesny_user['user_id']]
wczesny_user_tweety
```

```python
#5. Wyświetl tweety użytkownika. który ma najwięcej followersów.
max_followers_user = df.loc[df['user_followers_count'].idxmax()]
max_followers_tweets = df[df['user_id'] == max_followers_user['user_id']]
max_followers_tweets
```

```python
#6. Wyświetl tylko zweryfikowanych użytkowników (kolumna user_verified)
zweryfikowani_tweets = df[df['user_verified'] == True]
zweryfikowani_tweets
```

```python
#7.Podaj, w który dzień tygodnia najczęściej były publikowane tweety z datasetu
lista_dni = ['Poniedziałek', 'Wtorek', 'Środa','Czwartek', 'Piątek', 'Sobota', 'Niedziela']
count_dni = {dzien: df['created_at'].str.contains(dzien, na=False).sum() for dzien in lista_dni}
top_dzien = max(count_dni, key=count_dni.get)
top_dzien
```

```python
#Część 3
#Rozpoznaj osoby (persName) i przekaż je do kolumny o nazwie persons
persName_list = []
for item in df['text']:
   doc = nlp(item)
   for ent in doc.ents:
    if ent.label_ == "persName":
        persName_list.append(ent.text)
df['persons'] = persName_list + [None] * (len(df) - len(persName_list))
df
```

```python
#Rozpoznaj miejsca (placeName) i przekaż je do kolumny o nazwie places
placeName_list = []
for item in df['text']:
   doc = nlp(item)
   for ent in doc.ents:
    if ent.label_ == "placeName":
        placeName_list.append(ent.text)
df['places'] = placeName_list[:len(df)]
df
```

```python
#Rozpoznaj organizacje (orgName) i przekaż je do kolumny o nazwie organizations
orgName_list = []
for item in df['text']:
   doc = nlp(item)
   for ent in doc.ents:
    if ent.label_ == "orgName":
        orgName_list.append(ent.text)
df['organizations'] = orgName_list + [None] * (len(df) - len(orgName_list))
df
```
